```python
def preprocess_axis(image_path):
  path_to_amount_folder = r"/content/internshipATB/Dataset/axis/segmented amount"
  path_to_date_folder = r"/content/internshipATB/Dataset/axis/segmented date"
  path_to_word_folder = r"/content/internshipATB/Dataset/axis/segmented words"
  path_to_client_folder = r"/content/internshipATB/Dataset/axis/segmented client"
  path_to_id_folder = r"/content/internshipATB/Dataset/axis/segmented id"


  image = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB)
  (b, g, r) = image[100, 1000]
    # trying to change the pixels values to that colore
  for i in range(0, 200):
        for j in range(0, 1820):
            image[i, j] = (b, g, r)
    # Remove OR BEARER
  (b2, g2, r2) = image[500, 750]
  for i in range(240, 350):
        for j in range(1880, 2365):
            image[i, j] = (b2, g2, r2)
  for i in range(0, 83):
        for j in range(1750, 2340):
            image[i, j] = (b, g, r)
    # Attemping to remove the line under "RUPEES line" without losing data details
  (b3, g3, r3) = image[500, 750]
  for i in range(460, 540):
        for j in range(0, 1685):
            image[i, j] = (b3, g3, r3)
  for i in range(350, 400):
        for j in range(100, 380):
            image[i, j] = (b3, g3, r3)
  scale_percent = 100  # percent of original size
  width = int(image.shape[1] * scale_percent / 100)
  height = int(image.shape[0] * scale_percent / 100)
  dim = (width, height)
  image = cv2.cvtColor(cv2.resize(image, dim), cv2.COLOR_BGR2GRAY)
  image = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,  231, 14)
  image = cv2.bilateralFilter(image, 5, 75, 75)
  image = cv2.medianBlur(image,5)
  id_image = image[900:1200, 400:2090]
  image = image[40:530, 200:2285]
  words_image = image[270:470, 0:1490]
  amount_image = image[320:500, 1450:2225]
  client_image = image[70:300, 0:1490]
  date_image = image[0:200, 1560:2340]

  save_path_amount = os.path.join(path_to_amount_folder ,os.path.splitext(os.path.basename(image_path))[0] + ".png")
  cv2.imwrite(save_path_amount, amount_image)
  save_path_date = os.path.join(path_to_date_folder, os.path.splitext(os.path.basename(image_path))[0]+ ".png")
  cv2.imwrite(save_path_date, date_image)
  save_path_words = os.path.join(path_to_word_folder, os.path.splitext(os.path.basename(image_path))[0]+ ".png")
  cv2.imwrite(save_path_words, words_image)
  save_path_id = os.path.join(path_to_id_folder, os.path.splitext(os.path.basename(image_path))[0]+ ".png")
  cv2.imwrite(save_path_id, id_image)
  save_path_client = os.path.join(path_to_client_folder,os.path.splitext(os.path.basename(image_path))[0]+ ".png")
  cv2.imwrite(save_path_client, client_image)

```

```python
dataset_path = "/content/internshipATB/Dataset/300"
Axis_df = pd.read_csv('/content/AXIS BANK_df.csv')
cheque_column = 'CHEQUE NO '

# Parcourir chaque ligne de votre DataFrame
for index, row in Axis_df.iterrows():
    cheque_number = row[cheque_column]
    file_name = f"Cheque {cheque_number}.tif"  # Nom du fichier correspondant au numéro de chèque
    image_path = os.path.join(dataset_path, file_name)

    preprocess_axis(image_path)  # Appel de votre fonction preprocess_axis pour traiter l'image
```
